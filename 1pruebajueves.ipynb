Problema 4:
Un estudio medioambiental sugiere que el nivel promedio de monóxido carbono en el aire depende de la cantidad de habitantes, según la siguiente tabla:
Monóxido de carbono, partículas por millón (ppm)	1,05	1,45	1,5	1,75
Cantidad de habitantes	100	900	1000	1500


a)	Defina la variable dependiente e independiente, indicando unidad de medida.

b)	Determine la forma algebraica de la función lineal que modela la situación.


a)
variable independiente : cantidad habitantes (unidades) 
variable dependiente : nivel promedio de monocido carbono (ppm)
b)
obtencion forma akgebraia


```python
import numpy as np
vi = np.array([100, 900 , 1000,   1500])
vd = np.array([1.05, 1.45, 1.5,1.75])
pendiente, intercepto = np.polyfit(vi, vd, 1)
m=round(pendiente,7)
n=round(intercepto)
print(f"la funcion que modela la situacion es f(x)={m}x+{n}")
```

Problema 6:
El departamento de Recursos Humanos de una empresa, determinó que el bono de Navidad para el año 2016 dependerá de los años de antigüedad y del cargo que desempeña cada empleado. 
Utilizará las siguientes funciones:
Bono Administrativo: A(t)=4t+3t^2
Bono Cargos Superiores: S(t)=16+26t

Donde t es el número de años completos de antigüedad de los trabajadores, y las funciones determinan el valor del bono a pagar en miles de pesos según el cargo.


	a)¿Cuántos años de antigüedad se debe tener para que el monto del bono sea el mismo, independiente del cargo que ocupe la persona?


	b)Si los Administrativos recibieron un bono de 95 mil pesos, ¿Qué monto recibieron los Cargos superiores para esa misma cantidad de años de antigüedad?
	Interprete la pendiente del modelo lineal


```python
a)
bono admin : A(x)=4*t+3*t**2
```

```python
import sympy as sp
def A(t):
    return 4*t+3*t**2
def S(t):
    return 16+26*t

x=sp.Symbol("x")
ecuacion=A(x)-S(x)
solucion=sp.solve(ecuacion,x)
x1=round(solucion[0],5)
x2=round(solucion[1],5)
print(x1,x2)
print(f"a los 8 años de antiguedad los trabajadores reciben el mismo bono")
```
