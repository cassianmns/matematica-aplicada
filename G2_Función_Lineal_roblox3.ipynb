# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


```python
#respuesta 1
#dependiente = tiempo de ejecuccion (ms)
#independiente = Cantidad de elementos (unidades)

#respuesta 2

import numpy as np

vi=[100,200,500,1000,2000]
vd=[2,4,10,20,40]

coeficientes=np.polyfit(vi,vd,1)
print(np.around(coeficientes,5))

#Una forma mejor:

m1,n1=np.polyfit(vi,vd,1)
m=round(m1,5) #pendiente 
n=round(n1,5) #coeficiente de posicion 

print(m,n) #recordar que la forma algebraica de la funcion lineal es F(x)=mx+n
print(f"la forma algebraica es F(x)={m}x+{n}")
```

```python
#3 Calcula el tiempo de ejecuccion para una entrada de $1500 elementos.

def f(x):
    f=0.02*x+0
    return f
print(f"el tiempo de ejecucion para una entrada de 1500 elementos es de {f(1500)} ms")
```

```python
#4 utiliza la funcion solve de sympy para determinar el tamaño de entrada 
import sympy as sp
x=sp.Symbol("x")
ecuacion=f(x)-50
solucion=sp.solve(ecuacion,x)
print(solucion[0])
print(f"para un tiempo de 50 ms, el tamaño de la entrada debe ser {round(solucion[0])}")

#5 verifique el valor de la pendiente . 
#m=0,02 "El incremento en la funcion por el incremento en 1 unidad de x"
#el valor de la pendiente indica que por cada elemento el tiempo de ejecucion se incrementa en 0,02.
```

# Problema 10

Dos atletas profesionales corren por un camino recto. Durante los primeros $10$ segundos del trayecto, la distancia recorrida (en metros) por el atleta $1$ está dada por la función:

$$g(t) = 0,7t^{2}$$

mientras que la distancia recorrida por el atleta $2$ está dada por la función:

$$f(t) = 10t $$

donde $t$ corresponde al tiempo transcurrido (en segundos) desde el inicio del trayecto.

<br>

1. ¿Cuál de los dos atletas siempre mantuvo una velocidad constante? Justifique. *Recomendación: grafica las funciones utilizando Matlplotlib.*
2. Si la meta se encontraba a $100$ metros de la partida ¿cuál atleta llega primero?
3. ¿A qué distancia se encontraban uno del otro cuando transcurren $8$ segundos? ¿Y cuando transcurren $10$ segundos?
4. Entre el segundo $8$ y el $10$, ¿cuál atleta iba más rápido? Justifique apropiadamente.

```python
#variable dependiente = tiempo (t) (segundos)
#variable independiente = distancia (g - f) (metros)

import matplotlib.pyplot as plt
import numpy as np

def G(x):
    G=0.7*x**2
    return G #atleta 1
def F(x):
    F=10*x
    return F #atleta 2

x = np.arange(0,11,1)
plt.plot(x, G(x), label = 'Atleta 1 (G)')
plt.plot(x,F(x),label = "Atleta 2 (F)"  )
plt.title('Distancia en funcion del tiempo')
plt.ylabel('Tiempo en segundos')
plt.xlabel('Metros recorridos')
plt.legend()
plt.grid()
#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)
#

plt.show()
print("El atleta 2 siempre mantiene una velocidadd constante (10m/s)")
```

```python
#3
print(f"La distancia a los 8 segundos es {F(8)-G(8)} metros")
print(f"La distancia a los 8 segundos es {F(10)-G(10)} metros")
```
