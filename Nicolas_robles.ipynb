Imagina que estás trabajando en el departamento de monitoreo de servidores de una empresa de desarrollo de software. Tu equipo está analizando el desempeño de dos servidores que procesan solicitudes de usuarios durante el día.



El servidor A procesa solicitudes de acuerdo con la función ﻿A paréntesis izquierdo t paréntesis derecho igual 0 coma 2 t al cuadrado más 3 t más 5 coma 2﻿, donde ﻿A paréntesis izquierdo t paréntesis derecho﻿ representa la cantidad de solicitudes procesadas por este servidor después de ﻿t﻿ días.

El servidor B, por su parte, procesa solicitudes según la función ﻿B paréntesis izquierdo t paréntesis derecho igual 0 coma 2 t al cuadrado más 4 t más 3 coma 2﻿, donde ﻿B paréntesis izquierdo t paréntesis derecho﻿ representa la cantidad de solicitudes procesadas por este servidor después de ﻿t﻿ días.



Encuentra el momento en que ambos servidores habrán procesado la misma cantidad de solicitudes, resolviendo la ecuación ﻿A paréntesis izquierdo t paréntesis derecho igual B paréntesis izquierdo t paréntesis derecho﻿.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def A(x):
    return 0.2*x**2+3*x+5.2
def B(x):
    return 0.2*x**2+4*x+3.2

x=sp.Symbol("x")
ecuacion= A(x)-B(x)
solucion=sp.solve(ecuacion,x)
#print(solucion)
print(round(solucion[0]))

```

Una empresa obtiene ingresos por suscriptores según la función ﻿I paréntesis izquierdo x paréntesis derecho igual 60 x más 3 x al cuadrado﻿ y sus costos operativos se modelan con la función ﻿C paréntesis izquierdo x paréntesis derecho igual 50 x más 2400﻿ (ambos en dólares), donde ﻿x﻿ representa la cantidad de suscriptores.



Si los costos operativos alcanzan los 4.400 dólares, ¿cuál es el ingreso correspondiente?.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def I(x):
    return 60*x+3*x**2
def C(x):
    return 50*x+2400 

x=sp.Symbol("x")
ecuacion= I(x)-C(x)
solucion=sp.solve(ecuacion,x)
#print(solucion)
#print(round(solucion[0],3))


x = np.arange(0,100,1)
plt.plot(x, I(x), label = 'Ingreso suscripcion cantidad de segidores')
plt.plot(x, C(x), label = 'Costos operativos')
plt.title('Nombre del gráfico')
plt.ylabel('Nombre eje y')
plt.xlabel('Nombre eje x')
plt.legend()
plt.grid()
#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)
#

plt.show()
```

El tamaño de una base de datos aumenta un 5% semanalmente. Si inicialmente su tamaño es de 2 GB, el modelo que permitiría determinar el tamaño de la base de datos (en GB), transcurridas ﻿n﻿ semanas es:



﻿T paréntesis izquierdo n paréntesis derecho igual 2 por abrir paréntesis 1 coma 05 cerrar paréntesis elevado a n﻿



¿Cuál debería ser el tamaño de la base de datos al transcurrir 9 semanas?

```python
def T(x):
    return 2*(1.05)**x

print(round(T(9),0))
```

En el sitio de internet 1, donde no cobran gastos de envío, ofrecen un tablet en 320 dólares, mientras que en el sitio de internet 2 ofrecen el mismo tablet en 260 dólares, considerando en este caso un gasto fijo de envío y de internación de 180 dólares. Las funciones que permiten calcular el costo total (en dólares) en cada sitio de internet respectivamente, en función de la cantidad ﻿x﻿ de tablets comprados son ﻿f paréntesis izquierdo x paréntesis derecho igual 320 x﻿ y ﻿g paréntesis izquierdo x paréntesis derecho igual 260 x más 180﻿. El máximo de compras permitidas en cada sitio es de 10 unidades de tablets.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(x):
    return 320*x
def g(x):
    return 260*x+180


x=sp.Symbol("x")
ecuacion=f(x)-g(x) 
solucion=sp.solve(ecuacion,x)
print(solucion)
#print(round(solucion[0],3))

x = np.arange(0,15,1)
plt.plot(x, f(x), label = 'Sitio internet 1')
plt.plot(x, g(x), label = "Sitio internet 2")
plt.title('Tablets por internet')
plt.ylabel('Precio (dolares)')
plt.xlabel('Cantidad a comprar (unidades)')
plt.legend()
plt.grid()
#opcional:
ax=plt.gca()
ax.spines[["left","bottom"]].set_position(("data",0))
ax.spines[["right","top"]].set_visible(False)
#

plt.show()

print(f"En el sitio de internet 1, las 10 tablet salen con un precio de {f(10)} dolares")
print(f"en el sitio de internet 2, las 10 tablets salen a un precio de {g(10)} dolares")
print(f"sale mas conveniente comprar en el sitio 2 despues de las {solucion} unidades")
```
